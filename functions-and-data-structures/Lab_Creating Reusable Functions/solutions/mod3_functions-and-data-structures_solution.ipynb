## 1) Defining and calling functions

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Candace"))
print(greet("World"))
```

## 2) Functions with multiple parameters

```python
def calculate_total(price, quantity, tax_rate=0.07):
    subtotal = price * quantity
    tax = subtotal * tax_rate
    return subtotal + tax

total = calculate_total(price=19.99, quantity=3)
print("Total:", total)
```

## 3) Lists
Ordered collections of items.

```python
scores = [88, 92, 79, 85, 90]

print("Scores:", scores)
print("First score:", scores[0])
print("Number of scores:", len(scores))
```

## 4) Looping over a list

```python
total = 0

for score in scores:
    total += score

average = total / len(scores)
print("Average score:", average)
```

## 5) Dictionaries
Key-value pairs for structured data.

```python
student = {
    "name": "Alex",
    "age": 21,
    "major": "Data Science"
}

print("Student name:", student["name"])
print("Keys:", student.keys())
print("Values:", student.values())
```

## 6) Looping over dictionaries

```python
for key, value in student.items():
    print(key, "->", value)
```

## 7) Combining functions and data structures

```python
def average_scores(score_list):
    return sum(score_list) / len(score_list)

print("Average via function:", average_scores(scores))
```

## 8) Common pitfalls

- Forgetting to return a value from a function
- Modifying a list inside a loop without intending to
- Assuming dictionary keys are ordered (they are insertion-ordered in modern Python, but logic should not rely on it)
