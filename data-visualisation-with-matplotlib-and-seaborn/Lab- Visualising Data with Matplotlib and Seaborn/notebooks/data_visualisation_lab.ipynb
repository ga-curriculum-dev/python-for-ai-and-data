## 3) Matplotlib basics (clarity first)

### 3.1 Histogram — distribution
Create a histogram for a numeric column (e.g., `revenue_usd`).

✅ Requirements:
- Title
- Axis labels
- Appropriate binning (start simple)


```python
# TODO: choose your numeric column
col = "revenue_usd"  # change if needed

plt.figure(figsize=(8, 4))
plt.hist(df[col].dropna(), bins=10)
plt.title(f"Distribution of {col}")
plt.xlabel(col)
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

### 3.2 Bar chart — comparison by group

Create a bar chart that compares an aggregate metric by a category (e.g., total revenue by `category`).

✅ Requirements:
- Grouping + aggregation in Pandas
- Sorted bars (largest first)
- Clear labels


```python
# TODO: choose grouping column + metric
group_col = "category"     # e.g., category, state
metric_col = "revenue_usd" # e.g., revenue_usd, units_sold

summary = (
    df.groupby(group_col, dropna=False)[metric_col]
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(8, 4))
plt.bar(summary.index.astype(str), summary.values)
plt.title(f"Total {metric_col} by {group_col}")
plt.xlabel(group_col)
plt.ylabel(f"Total {metric_col}")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()

summary
```

### 3.3 Scatter plot — relationship between two numeric variables

Create a scatter plot to explore a relationship (e.g., `price_usd` vs `units_sold`).

✅ Requirements:
- Title
- Axis labels
- Optional: add transparency (`alpha`) if many points


```python
x = "price_usd"
y = "units_sold"

plt.figure(figsize=(6, 4))
plt.scatter(df[x], df[y], alpha=0.8)
plt.title(f"{y} vs {x}")
plt.xlabel(x)
plt.ylabel(y)
plt.tight_layout()
plt.show()
```

## 5) Apply visual design principles (refine at least one chart)

Pick **one** chart you already created and refine it.

In the cell below, refine your **bar chart** (example). You can refine a different chart if you prefer.


```python
# Example refinement: bar chart with formatting tweaks
summary = (
    df.groupby("category", dropna=False)["revenue_usd"]
      .sum()
      .sort_values(ascending=False)
)

plt.figure(figsize=(8, 4))
bars = plt.bar(summary.index.astype(str), summary.values)
plt.title("Total Revenue (USD) by Category")
plt.xlabel("Category")
plt.ylabel("Total Revenue (USD)")
plt.xticks(rotation=30, ha="right")

# Optional: add value labels (use sparingly)
for i, v in enumerate(summary.values):
    plt.text(i, v, f"{v:,.0f}", ha="center", va="bottom", fontsize=9)

plt.tight_layout()
plt.show()
```
