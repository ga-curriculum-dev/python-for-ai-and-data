## 1) Setup and load data

```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd().parent
DATA_DIR = PROJECT_ROOT / "data"
OUTPUT_DIR = PROJECT_ROOT / "output"

csv_file = DATA_DIR / "sales_sample.csv"
df = pd.read_csv(csv_file)

df.head()
```

## 2) Inspect the data

## 3) Fix data types

```python
df["order_date"] = pd.to_datetime(df["order_date"])

df.dtypes
```

## 4) Create a calculated column

```python
df["revenue"] = df["quantity"] * df["unit_price"]

df.head()
```

## 5) Filter and sort
We exclude returned orders to keep revenue summaries accurate.

```python
df_complete = df[df["status"] == "Complete"].copy()
df_complete_sorted = df_complete.sort_values(by="revenue", ascending=False)

df_complete_sorted.head()
```

## 6) Group and summarize

```python
revenue_by_region = (
    df_complete
      .groupby("region", as_index=False)
      .agg(
          orders=("order_id", "nunique"),
          units=("quantity", "sum"),
          total_revenue=("revenue", "sum"),
          avg_order_value=("revenue", "mean")
      )
)

revenue_by_region
```

## 7) Create a pivot table

```python
pivot = pd.pivot_table(
    df_complete,
    values="revenue",
    index="region",
    columns="product",
    aggfunc="sum",
    fill_value=0
)

pivot
```

## 8) Export outputs

```python
df_complete_sorted.to_csv(OUTPUT_DIR / "sales_cleaned_complete.csv", index=False)
revenue_by_region.to_csv(OUTPUT_DIR / "revenue_by_region.csv", index=False)
pivot.to_csv(OUTPUT_DIR / "revenue_pivot.csv")
```
