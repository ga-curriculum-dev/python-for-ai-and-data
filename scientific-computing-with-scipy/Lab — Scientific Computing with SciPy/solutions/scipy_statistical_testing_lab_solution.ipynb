## 1) Setup & Environment Check

```python
import numpy as np
import pandas as pd
import scipy
from scipy import stats
import matplotlib.pyplot as plt

print("NumPy version:", np.__version__)
print("SciPy version:", scipy.__version__)
```

## 2) Load the Dataset

```python
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent
DATA_DIR = PROJECT_ROOT / "data"

df = pd.read_csv(DATA_DIR / "experiment_results.csv")
df.head()
```

```python
print("Rows, cols:", df.shape)
print(df.dtypes)
print(df["group"].value_counts())
df.describe(include="all")
```

## 3) Compute Descriptive Statistics

```python
df["value"].describe()
```

```python
df.groupby("group")["value"].describe()
```

## 4) Define Your Hypothesis

### Hypothesis

**Null hypothesis (H₀):** The mean value is the same for the control and treatment groups.  
**Alternative hypothesis (H₁):** The mean value differs between the control and treatment groups.


## 5) Perform a Hypothesis Test (SciPy)

```python
control = df[df["group"] == "control"]["value"].to_numpy()
treatment = df[df["group"] == "treatment"]["value"].to_numpy()

control[:5], treatment[:5]
```

```python
# Welch's t-test is a safer default when group variances may differ.
t_stat, p_value = stats.ttest_ind(control, treatment, equal_var=False)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

## 6) Visualise the Data

```python
plt.boxplot([control, treatment], labels=["Control", "Treatment"])
plt.ylabel("Value")
plt.title("Control vs Treatment")
plt.show()
```

## 7) Interpret and Document Results

### Results & Interpretation

A p-value tells you how surprising the observed difference would be **if the null hypothesis were true**.

A common threshold is 0.05:
- If **p < 0.05**, you would typically reject the null hypothesis.
- If **p ≥ 0.05**, you do not have enough evidence to reject the null hypothesis.

In practice, you should interpret statistical results alongside context, sample size, and practical significance (not just statistical significance).
