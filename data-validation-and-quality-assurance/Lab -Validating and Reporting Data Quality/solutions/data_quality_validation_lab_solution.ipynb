# Data Quality Validation Lab — Solution

```
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv('../data/dataset.csv')
```

```
df.info()
df.isna().sum()
df.duplicated().sum()
```

```
# Choose the ID column (this dataset uses `product_id`)
id_col = "id"
if id_col not in df.columns:
    # fallback: pick the first column that ends with "_id"
    id_candidates = [c for c in df.columns if c.lower().endswith("_id")]
    if id_candidates:
        id_col = id_candidates[0]

assert id_col in df.columns, f"Expected an ID column, but none found. Columns: {list(df.columns)}"
assert df[id_col].notna().all(), f"Missing values detected in {id_col}"

# Basic de-duplication and required fields
df = df.drop_duplicates()
# If a 'title' column doesn't exist in this dataset, skip that rule
if "title" in df.columns:
    df = df.dropna(subset=["title"])

df.head()
```

```
df.to_csv('../output/validated_dataset.csv', index=False)
df.head()
```
