## 1) Variables + core data types

```python
# Integers and floats
age = 34
temperature_f = 72.5

# Strings
name = "Candace"
city = "New York"

# Booleans
is_active = True
has_coupon = False

print("age:", age, type(age))
print("temperature_f:", temperature_f, type(temperature_f))
print("name:", name, type(name))
print("city:", city, type(city))
print("is_active:", is_active, type(is_active))
print("has_coupon:", has_coupon, type(has_coupon))
```

## 2) Arithmetic operators

```python
a = 10
b = 3

print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)      # division returns float
print("a // b =", a // b)    # floor division
print("a % b =", a % b)      # remainder (modulo)
print("a ** b =", a ** b)    # exponent
```

## 3) Comparison operators
These return booleans.

```python
x = 8
y = 12

print("x == y:", x == y)
print("x != y:", x != y)
print("x < y:", x < y)
print("x <= y:", x <= y)
print("x > y:", x > y)
print("x >= y:", x >= y)
```

## 4) Logical operators
Useful for combining conditions.

```python
is_member = True
cart_total = 47.00
free_shipping_threshold = 50.00

eligible_for_free_shipping = is_member or cart_total >= free_shipping_threshold
eligible_for_discount = is_member and cart_total >= 25.00

print("eligible_for_free_shipping:", eligible_for_free_shipping)
print("eligible_for_discount:", eligible_for_discount)

# not flips a boolean
print("not is_member:", not is_member)
```

## 5) Conditional logic with if / elif / else

```python
score = 83

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print("Score:", score)
print("Grade:", grade)
```

## 6) Edge cases: order matters
Demonstrates why condition ordering changes results.

```python
value = 5

# Incorrect ordering (example)
if value >= 0:
    label = "non-negative"
elif value > 10:
    label = "greater than 10"
else:
    label = "negative"

print("Incorrect ordering label:", label)

# Correct ordering
if value > 10:
    label = "greater than 10"
elif value >= 0:
    label = "non-negative"
else:
    label = "negative"

print("Correct ordering label:", label)
```

## 7) Quick debugging habits
A tiny checklist you can reuse.

When code behaves strangely:
- Print the value you *think* you have.
- Print the type (`type(x)`) when math/logic is off.
- Simplify the condition into smaller parts.
- Re-run cells top to bottom if state feels inconsistent.


```python
# Example: debugging a condition
user_input = "25"   # note: this is a string, not an int

print("user_input:", user_input, type(user_input))

# Fix by converting
age = int(user_input)
print("age:", age, type(age))

if age >= 21:
    print("Access granted.")
else:
    print("Access denied.")
```
