## 2) Check Python details

```python
import sys
import platform

print("Python version:", sys.version.split()[0])
print("Executable:", sys.executable)
print("Platform:", platform.platform())
```

## 3) Verify common data libraries
If these imports fail, install the missing packages in your environment and re-run.

```python
# Common libraries used in later modules
imports = ["numpy", "pandas", "matplotlib"]

for name in imports:
    try:
        __import__(name)
        print(f"✅ {name} imported successfully")
    except Exception as e:
        print(f"❌ {name} import failed:", e)
```
